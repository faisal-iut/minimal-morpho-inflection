```python
from log import logger
from itertools import groupby, combinations
from operator import itemgetter
from nltk import FreqDist
from collections import defaultdict
from sklearn.metrics.pairwise import cosine_similarity

# print(get_prefix(list_of_strings, 2))
# print(get_prefix(list_of_strings, 6))
def get_affix(strings):
    suffixDist=FreqDist()
    prefixDist=FreqDist()
    strings = [s for s in strings if type(s)!=int and len(s)>1]
    for word in strings:
        for i in range(1,4):
            prefixDist[word[:i]] +=1
            suffixDist[word[-i:]] +=1
    s3=[]
    s2=[]
    s1=[]
    cs = suffixDist.most_common(100)
    cp = prefixDist.most_common(100)
    dist_dict ={}
    for i in range(2,4):
        dist_dict['s{}'.format(i)] = [(s,c) for s,c in cs if len(s)==i][:20]
        dist_dict['p{}'.format(i)] = [(p,c) for p,c in cp if len(p)==i][:20]
    return dist_dict

def connected_components(L):
    pool = set(map(frozenset, L))
    groups = []
    while pool:
        group = set()
        groups.append([])
        while True:
            for candidate in pool:
                if not group or group & candidate:
                    group |= candidate
                    groups[-1].append(tuple(candidate))
                    pool.remove(candidate)
                    break
            else:
                break
    return groups

def merge_table(df,sm):
    col_2_id = dict(zip(df.columns,range(len(df.columns))))
    id_2_col = dict(zip(range(len(df.columns)),df.columns))
    rc = [c for c in combinations(col_2_id.values(),2) if 
          all((df[id_2_col[c[0]]]==0) | (df[id_2_col[c[1]]]==0))]
    if len(rc)==0:
        return [],df, False
    ccomp = connected_components(rc)
    merged_pairs = []
    for cc in ccomp:
        rc_sim = [(c, sm[c[0]][c[1]]) for c in cc]
        rc_sim = sorted(rc_sim,key=itemgetter(1))
        p2m1, p2m2 = id_2_col[rc_sim[-1][0][0]], id_2_col[rc_sim[-1][0][1]]
        merged_pairs.append((p2m1,p2m2))
        col = "{},{}".format(p2m1,p2m2)
        df[col] = df[p2m1]
        df.loc[df[col]==0,col]=df[p2m2]
        df = df.drop(columns =[p2m1,p2m2])
    return merged_pairs, df, True

def save_current_state(df, lemma):
    df_np = df.values
    coun=0
    my_col_2_wf={}
    my_wf_2_col={}
    my_base_2_wf={}
    my_wf_2_base={}
    for i in df_np:
        each_lemma  = '{}_{}'.format(coun,lemma[coun])
        my_base_2_wf[each_lemma] = set()
        for k in range(len(i)):
            if i[k]!=0:
                if k not in my_col_2_wf:
                    my_col_2_wf[k] = set()
                my_col_2_wf[k].add(i[k])
                my_wf_2_col[i[k]]={k}
                my_base_2_wf[each_lemma].add(i[k])
                my_wf_2_base[i[k]]={each_lemma}
#         print("%15s %15s %15s %15s %15s %15s"%(i[0],i[1], i[2],i[3], i[4],i[5]))
        coun = coun+1
#     print("%15s"%(i[2]))

    my_data = {'col_2_wf': my_col_2_wf,
              'wf_2_col': my_wf_2_col,
              'base_2_wf': my_base_2_wf,
              'wf_2_base': my_wf_2_base,
                'r': df_np.shape[0],
                'c': df_np.shape[1]}

    with open(os.path.join(pfilepath, 'my_data_#cell_{}.pickle'.format(df_np.shape[1])),'wb') as ff:
        pickle.dump(my_data,ff)
    logger.info("state {} saved at {}".format(df_np.shape[1], pfilepath))
    
def get_sim_matrix(df):
    affix_count = defaultdict(dict)
    for col in df.columns:
        dist_dict = get_affix(set(df[col]))
        for key,val in dist_dict.items():
            for v in val:
#                 affix_count[v[0]]['affix']=key[0]
                affix = '{}_{}'.format(key[0],v[0])
                affix_count[affix][col]=v[1]
    affix_df = pd.DataFrame.from_dict(affix_count, orient='index')
    affix_df = affix_df.fillna(0)
    affix_np = affix_df.T.values
    sim_matrix = cosine_similarity(affix_np)
    return sim_matrix, affix_df

if __name__ == "__main__":
    filepath = './data/eng_V/'
    with open(os.path.join(filepath, 'dst_property_t.pickle'),'rb') as ff:
        dst_verbs_t = pickle.load(ff)
    start_col = 6
    feature_df = dst_verbs_t.iloc[:,start_col:].copy()
    lemma =  list(dst_verbs_t.iloc[:,1])
    for i in range(0,len(feature_df.columns)):
        sim_matrix, affix_df = get_sim_matrix(feature_df.copy())
        mp, new_table, flag = merge_table(feature_df.copy(),sim_matrix)
#         logger.info("merged columns: {}".format(mp))
        logger.info("# before: {:3d}, # after: {:3d}".format(feature_df.shape[1],new_table.shape[1]))
#         logger.info("\n\n")
        save_current_state(feature_df.copy(),lemma)
        if flag==False:
            break
        feature_df = new_table.copy()
    feature_df.to_csv('./data/my_wf_grid.csv',header=None)
```
